```python
import os
import pandas as pd
import src
```

```python
result_path = src.reports_dir / "confusion_matrices"
```

```python
def accuracy(tp, fn, fp, tn):
    return (tp + tn) / (tp + fn + fp + tn)
def recall(tp, fn, fp, tn):
    return tp / (tp + fn)
def precision(tp, fn, fp, tn):
    if tp + fp > 0:
        return tp / (tp + fp)
    else:
        return 0
def f1_score(tp, fn, fp, tn):
    p, r = precision(tp, fn, fp, tn), recall(tp, fn, fp, tn)
    return 2 * (p * r) / (p + r)


def attach_measures(r):
    res_tp = r.copy()
    res_tp['accuracy'] = res_tp.iloc[:, :4].apply(lambda x: accuracy(*x), axis=1)
    res_tp['recall'] = res_tp.iloc[:, :4].apply(lambda x: recall(*x), axis=1)
    res_tp['precision'] = res_tp.iloc[:, :4].apply(lambda x: precision(*x), axis=1)
    res_tp['f1-score'] = res_tp.iloc[:, :4].apply(lambda x: precision(*x), axis=1)
    return res_tp
```

```python
ress = []
for tissue in os.listdir(result_path):
    print(tissue)
    tissue_path = result_path / tissue
    for pipeline in os.listdir(tissue_path):
        print("\t{}".format(pipeline.split(".")[0]))
        pipeline_path = tissue_path / pipeline
        res_tp = pd.read_table(pipeline_path, index_col=0)
        res_tp_measures = attach_measures(res_tp)
        res_tp = res_tp_measures
        stds = res_tp_measures.iloc[:, 6:].std(0).to_frame().T
        stds.columns = stds.columns.map(lambda x: x + '_std')
        
        res_tp['tissue'] = tissue
        res_tp['method'] = pipeline.split(".")[0]
        ress.append(res_tp)
ress = pd.concat(ress, axis=0, ignore_index=True)#.iloc[:, [12, 13, 4, 5, 6, 7, 8, 9, 10, 11]]
ress
```

```python
import numpy as np
import utils
import pandas as pd
```

```python
r = []
for tn in utils.all_tumor_names:
    print(tn)
    X_tn, y_tn = utils.get_cancer_data(cancer_name=tn)
    n_samples = X_tn.shape[0]
    n_genes = X_tn.shape[1]
    n_tumor, n_normal = y_tn[y_tn == utils.TUMOR].shape[0], y_tn[y_tn == utils.NORMAL].shape[0]
    
    r.append({
        'cohort': tn,
        '#samples': n_samples,
        '#genes': n_genes,
        '#tumor': n_tumor,
        '#normal': n_normal
    })
r = pd.DataFrame.from_dict(r).set_index("cohort")[['#samples', '#tumor', '#normal', '#genes']]
r
```

```python
r.to_excel("./tcga_meta.xlsx", index=True, header=True)
```
